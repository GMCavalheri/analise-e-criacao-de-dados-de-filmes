```python
import pandas as pd
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

def recomendacao_apriori(dados, suport_min = 0.01, confi_min = 0.1, lift_min = 0.5):
    dados_atributos = dados.drop(columns='id_consumidor')
    dados_if = apriori(dados_atributos,min_support=suport_min,use_colnames=True)   
    dados_ar = association_rules(dados_if, metric='confidence', min_threshold=confi_min)
    dados_selecionados = dados_ar[dados_ar['lift']>lift_min]
    return dados_selecionados[['antecedents', 'consequents']]
```

```python
dados = pd.read_csv('banco_de_dados_relacional_de_filmes.csv')

recomendacao_apriori(dados)
```
